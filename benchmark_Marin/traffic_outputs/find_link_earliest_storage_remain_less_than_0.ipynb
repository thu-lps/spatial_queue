```python
data1 = pd.DataFrame()
for link_id in [39748, 37947, 13707, 37841, 39708, 31859, 37190, 20534, 18737, 3217, 37492, 31486, 10866]:
    for i in range(1000, 48000, 1000):
        path = './fairfax/with_bg/link_stats/lfairfax_vect_at_{}_bg_1_real_departure_time.csv'.format(i)
        data = pd.read_csv(path)
        data['link_id'] = data['link_id'].astype('str')
        data['head'] = data['link_id'].apply(lambda x: x[0])
        data = data[data['head']!= 'v']
        data = data[['link_id', 'queue', 'run', 'undeparted', 'length', 'storage_remain', 'geometry']]
        data['link_id'] = data['link_id'].astype('int')
        a = data.loc[data['link_id'] == link_id,:].copy()
        if len(a) == 0: #it means there is no car there
            continue;
        else:
            a.reset_index(drop = True, inplace = True)
            a.loc[0,'time'] = i
            data1 = data1.append([a], ignore_index = True)
```
