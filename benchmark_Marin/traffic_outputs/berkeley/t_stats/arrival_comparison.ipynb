```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = './arrival_berkeley_total.csv'
t_df_sq = pd.read_csv(path)

path = './t_stats_microsim.csv'
t_df_micro = pd.read_csv(path)
```

```python
plt.figure()
plt.plot(t_df_sq['t'], t_df_sq['arrival_count'], 'r--', label = 'spatial queue')
plt.plot(t_df_micro['t'], t_df_micro['arrival_count'], 'b--', label = 'microsim')
plt.xlabel('Travel time [s]')
plt.ylabel('Number of evacuees that reach the destination')
plt.legend()
plt.tight_layout()
plt.savefig('./comparison_between_spatial_queue_and_microsim.jpg', dpi =800)
plt.show()
```
