```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
directory = './speed_data_from_Google'
weekday_list = ['Saturday', 'Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
weekday_map = {'Saturday':0,'Sunday':1,'Monday':2,'Tuesday':3,'Wednesday':4,'Thursday':5,'Friday':6}
speed_route_df = pd.DataFrame()
```

```python
for weekday in weekday_list:
    for hour in range(0,24):
        path = directory + '/' + weekday + '_' + str(hour) + '_speed.csv'
        data = pd.read_csv(path)
        speed_route_df = pd.concat([speed_route_df, data])
```

```python
speed_route_df['week_num'] = speed_route_df['week'].map(weekday_map)
speed_route_df['rid'] = speed_route_df['RouteID'].apply(lambda x: int(x[-1]))
```

```python
speed_route_df.sort_values(by = ['rid', 'week_num'], inplace = True)
```

```python
speed_route1 = speed_route_df.loc[speed_route_df['rid'] == 1]
speed_route1.reset_index(drop = True, inplace = True)

speed_route2 = speed_route_df.loc[speed_route_df['rid'] == 2]
speed_route2.reset_index(drop = True, inplace = True)

speed_route3 = speed_route_df.loc[speed_route_df['rid'] == 3]
speed_route3.reset_index(drop = True, inplace = True)

speed_route4 = speed_route_df.loc[speed_route_df['rid'] == 4]
speed_route4.reset_index(drop = True, inplace = True)

speed_route5 = speed_route_df.loc[speed_route_df['rid'] == 5]
speed_route5.reset_index(drop = True, inplace = True)

speed_route6 = speed_route_df.loc[speed_route_df['rid'] == 6]
speed_route6.reset_index(drop = True, inplace = True)

speed_route7 = speed_route_df.loc[speed_route_df['rid'] == 7]
speed_route7.reset_index(drop = True, inplace = True)

speed_route8 = speed_route_df.loc[speed_route_df['rid'] == 8]
speed_route8.reset_index(drop = True, inplace = True)

speed_route9 = speed_route_df.loc[speed_route_df['rid'] == 9]
speed_route9.reset_index(drop = True, inplace = True)
```

```python
speed_route1.loc[:,'sequence'] = np.arange(len(speed_route1))
speed_route2.loc[:,'sequence'] = np.arange(len(speed_route2))
speed_route3.loc[:,'sequence'] = np.arange(len(speed_route3))
speed_route4.loc[:,'sequence'] = np.arange(len(speed_route4))
speed_route5.loc[:,'sequence'] = np.arange(len(speed_route5))
speed_route6.loc[:,'sequence'] = np.arange(len(speed_route6))
speed_route7.loc[:,'sequence'] = np.arange(len(speed_route7))
speed_route8.loc[:,'sequence'] = np.arange(len(speed_route8))
speed_route9.loc[:,'sequence'] = np.arange(len(speed_route9))
```
